```python
from IPython.display import display, Math, Latex
from numpy import heaviside as H
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
# Данная программ предназначена для отображения волновой картины,
# которая реализуется в упругом стержне при рассмотрении одно-
# мерных волн.
```

```python
# Определим параметра
A = -6 # Амплитуда импульса
T_1 = 0.7 # Продолжительность участка роста напряжений
T_2 = 0.8 # Продолжительность участа спада напряжений
L = 5 # Длина стержня
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(0, L), ylim=(-10, 10))
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return line,
```

```python
# В данной программе рассматривается случай, при котором импульс
# напряжений имеет форму треугольника (с участками спада и
# и роста напряжений)
# Формула для напряжений выглядит следующим образом
display(Math(r"\sigma(t) = \dfrac{At}{T_{1}}  \left[H(t) - H(t-T_{1}) \right ] +\dfrac{A(T_{2}-t)}{T_{2}-T_{1}} \left[H(t-T_{1}) - H(t-T_{2}) \right ]"))
```

```python
def animate(i):
    t = np.linspace(0, 5, 1000)
    scale_parameter = -1.5
    #arg left -> right wave
    arg_lw = t - i*0.01 -scale_parameter
    # arg right -> left wave
    arg_rw = t+i*0.01 +scale_parameter
    y = (A / T_1 * (arg_lw) * (H(arg_lw, 0) - H(arg_lw - T_1 , 0)) + A / (T_2 - T_1) * (
    T_2 - arg_lw ) * (H(arg_lw - T_1 , 0) - H(arg_lw - T_2, 0))) - (
        (A / (T_2 - T_1) * (arg_rw - 2 * L + T_2)) * (
        H(arg_rw - 2 * L + T_2, 0) - H(arg_rw - 2 * L + T_1, 0)) + (A / T_1 * (-(arg_rw) + 2 * L)) * (
        (H(arg_rw - 2 * L + T_1, 0)) - H(arg_rw - 2 * L, 0)))
    line.set_data(t, y)
    return line,
anim = animation.FuncAnimation(
    fig, animate, init_func=init, frames=800, interval=30, blit=True)
```

```python
# Видео, демонстрирующее процесс отражения волны от свободной
# поверхности стержня
HTML(anim.to_html5_video())
```
